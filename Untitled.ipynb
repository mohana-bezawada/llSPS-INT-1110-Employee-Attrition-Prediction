# Labelencoding and onehotencoding

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
l = LabelEncoder() 
```

```python
print(df.shape)
```

```python
df.head()
```

```python
#Label Encoding will be used for columns having less than or equal 2 values
for i in df[1:]:
    if df[i].dtype == 'object':
        if len(list(df[i].unique())) <= 2:
            l.fit(df[i])
            df[i] = l.transform(df[i])
```

```python
#dummy values are for remaining categorial values
df = pd.get_dummies(df, drop_first = True)
```

```python
print(df.shape)
```

```python
df.head(5)
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,5))
col = list(df.columns)
col.remove('Attrition')
for i in col:
    df[i] = df[i].astype(float)
    df[[i]] = scaler.fit_transform(df[[i]])
df['Attrition'] = pd.to_numeric(df['Attrition'], downcast = 'float')
df.head()
```

```python
print(df.shape)
```
